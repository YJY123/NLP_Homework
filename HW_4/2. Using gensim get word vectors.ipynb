### 3. 导入相关模块

```python
import gensim
import multiprocessing
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
```

```python
import logging
logging.basicConfig(format="%(asctime)s:%(levelname)s:%(message)s",level=logging.INFO)
```

### 4. 训练模型并保存

```python
model = Word2Vec(LineSentence('/home/aistudio/data/data10734/wiki_corpus'), size=100, window=5, min_count=1, workers=4)
model.save("data/wiki_corpus.model")
```

### 5. 查看生成的词向量

```python
vector = model.wv['猪头'] 
```

```python
vector
```

### 6. 测试相似词

```python
one_corpus = ["人工智能"]
result = model.most_similar(one_corpus[0],topn=20)
```

```python
result
```

```python
two_corpus = ["腾讯","阿里巴巴"]
res = model.similarity(two_corpus[0],two_corpus[1])
print (res)
```
